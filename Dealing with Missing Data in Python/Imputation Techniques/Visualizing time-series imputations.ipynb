# Visualizing time-series imputations


```python
#import packages
import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read_data
airquality = pd.read_csv("../datasets/air-quality.csv" , parse_dates=['Date'] , index_col='Date')
airquality.head()
```

```python
plt.style.use('bmh')

airquality['Ozone'].plot(title="ozon" , marker="o" , figsize=(20,5))

```

### Ffill Imputation


```python
ffill_imp = airquality.fillna(method='ffill')
```

```python
ffill_imp['Ozone'].plot(color='red', marker='o', linestyle='dotted', figsize=(30, 5))
airquality['Ozone'].plot(title='Ozone', marker='o')
```

### Bfill Imputation


```python
bfill_imp = airquality.fillna(method='bfill')
bfill_imp['Ozone'].plot(color='red', marker='o', linestyle='dotted', figsize=(30, 5))
airquality['Ozone'].plot(title='Ozone', marker='o')
```

### Linear Interpolation


```python
linear_interep = airquality.interpolate(method='linear')
linear_interep['Ozone'].plot(color='red', marker='o', linestyle='dotted', figsize=(30, 5))
airquality['Ozone'].plot(title='Ozone', marker='o')
```

### Quadratic Interpolation


```python
quadratic_interep = airquality.interpolate(method='quadratic')
quadratic_interep['Ozone'].plot(color='red', marker='o', linestyle='dotted', figsize=(30, 5))
airquality['Ozone'].plot(title='Ozone', marker='o')
```

### Nearsest Interpolation


```python
nearest_interep = airquality.interpolate(method='nearest')
nearest_interep['Ozone'].plot(color='red', marker='o', linestyle='dotted', figsize=(30, 5))
airquality['Ozone'].plot(title='Ozone', marker='o')
```

## A comparison of the interpolations


```python
# Create subplots
fig, axes = plt.subplots(3, 1, figsize=(30, 20))
# Create interpolations dictionary
interpolations = {'Linear Interpolation': linear_interep,
                  'Quadratic Interpolation': quadratic_interep,
                  'Nearest Interpolation': nearest_interep}

# Visualize each interpolation
for ax, df_key in zip(axes, interpolations):
    interpolations[df_key].Ozone.plot(color='red', marker='o',linestyle='dotted', ax=ax)
    airquality.Ozone.plot(title=df_key + '- Ozone', marker='o', ax=ax)
```
