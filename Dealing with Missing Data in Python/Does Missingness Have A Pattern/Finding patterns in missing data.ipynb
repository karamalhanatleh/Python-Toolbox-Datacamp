# Finding patterns in missing data

```python
#import
import pandas as pd
import missingno as msno
import matplotlib.pyplot as plt
```

```python
#read_data
diabetes = pd.read_csv("../datasets/pima-indians-diabetes data.csv")
diabetes.head()
```

```python
msno.matrix(diabetes)
```

### Missingness Heatmap


<b>Graph of correlation of missing values between columns</b><br>

<b>Explains the dependencies of missingness between columns</b><br>

```python
msno.heatmap(diabetes )
```

### Missingness Dendrogram

<b>Tree diagram of missingness </b><br>

<b>Describes correlation of variables by grouping them</b><br>

```python
msno.dendrogram(diabetes)
```
